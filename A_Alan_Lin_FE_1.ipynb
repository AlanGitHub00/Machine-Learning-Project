```python
import numpy as np
import pandas as pd
outliers=[]
def detect_outlier(x):
    
    threshold=3
    mean_1 = np.mean(x)
    std_1 =np.std(x)
    
    
    for y in x:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```

```python
outlier_datapoints = detect_outlier(x)
print(outlier_datapoints)
```
