EX-1 - Mean / Median / STD

We have some sample salary data (in thousands) from two cities.

```python
city1 = [15,12, 20, 25, 50, 35, 75, 80, 60, 45, 36]
city2 = [40,42, 45, 60, 55, 52, 56, 52, 62, 57, 48]
```

Calculate mean, median, variation, standard deviation for both city data.

```python
import statistics
x1 = statistics.mean(city1)
print("The mean of city1 is", x1)        #mean for city1
```

```python
x2 = statistics.mean(city2)
print("The mean of city2 is", x2)    #mean for city2
```

```python
x3 = statistics.median(city1)
print("The median of city1 is", x3)  #median for city 1
```

```python
x4 = statistics.median(city2)
print("The median of city2 is", x4) #median for city 2
```

```python
import numpy as np
x5 = np.std(city1)
print("The standard deviation of city1 is",x5) #standard deviation of city1
```

```python
x6 = np.std(city2)
print("The standard deviation of city2 is",x6) #standard deviation of city2
```

```python
df = pd.read_csv('nba-player-stats.csv')

# drop rows with no height data
df = df.dropna(subset=['Height'])     

# split the strings
df.Height = [s.split('-') for s in df.Height]
# convert to inches
df.Height = [float(value[0])*12 + float(value[1]) for value in df.Height]

# write data
df.to_csv('nba-player-stats.csv', index=False)
```
