A - Encoding

A-1 : Index Encoding

1. Create the following data frame - Index encode (integer encode) status column

```python
import pandas as pd
df = pd.DataFrame({"customer_id":[1,2,3,4,5,6,7], 
                      "status":['married', 'single', 'single', 'divorced', 'married' ,'single', 'married' ]})
df
```

```python
df['status'] = df['status'].astype('category')
```

```python
df.dtypes
```

```python
df["status_cat"] = df["status"].cat.codes
df
```

A-2: One-Hot Encoding

One-hot encode status column in the above dataframe

```python
pd.get_dummies(df, columns=["status"])
```
